```python
from google_play_scraper import app
import pandas as pd
import numpy as np
```

```python
from google_play_scraper import Sort, reviews

result, continuation = reviews(
     'com.gojek.app',
     lang='id',
     country='id',
     sort=Sort.NEWEST,
     count=30000,
     filter_score_with=None
 )
```

```python
data = pd.DataFrame(np.array(result), columns=['review'])
data = data.join(pd.DataFrame(data.pop('review').tolist()))
data.head()
```

```python
data.shape
```

```python
# # Menambahkan kolom "Kota"
num_rows = data.shape[0]

# # Daftar kota di Jakarta
sumut = ['Medan', 'Kisaran', 'Pematang Siantar', 'Tebing Tinggi', 'Binjai']

# # Menambahkan kolom "Kota" dengan nilai acak
data['kota'] = np.random.choice(sumut, size=num_rows)

# # Tampilkan beberapa data
data.head()
```

```python
# Simpan data
data.to_csv('reviews_gojek.csv', index=False)
```
